```python
from google_play_scraper import Sort, reviews
from google_play_scraper import app
import pandas as pd
import numpy as np
```

```python
from google_play_scraper import app

result = app('com.snapchat.android')  # ID aplikasi dari URL Play Store

print("Nama Aplikasi:", result['title'])
print("Developer:", result['developer'])
print("Rating:", result['score'])
print("Jumlah Unduhan:", result['installs'])
print("Deskripsi Singkat:", result['summary'])
```

```python
result, continuation_token = reviews(
    'com.snapchat.android',
    lang='id', #bahasa
    country='id', #negara
    sort=Sort.MOST_RELEVANT, #urutan
    count=2000, #jumlah dataset
    filter_score_with=None #rating: none jika ingin campur, atau bisa isi dengan 1,2,3,4,5
)
```

```python
#Dataframe dengan nama
data = pd.DataFrame(np.array(result),columns=['review'])
data = data.join(pd.DataFrame(data.pop('review').tolist()))
data.head()
```

```python
#mengidentifikasi seberapa panjang data yang sudah di scraping
len(data)
```

```python
#mengambil fitur hanya content dan skor untuk kebutuhan selanjutnya
data = data[['content','score']]
data.head()
```

```python
data = data.rename(columns={'content': 'ulasan', 'score': 'label' }) #mengganti nama fitur
data.head()
```

```python
#mengekspor dataset
#bisa menemukannya pada directory google collab
data.to_csv("Ulasan Snapchat 2000 Data.csv", index = False , encoding='utf-8')
```
